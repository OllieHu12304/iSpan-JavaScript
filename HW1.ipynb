```python
# 1.	選擇性敘述的練習-season
# 輸入月份1~12月，利用switch判斷相對應的季節春、夏、秋、冬並印出。若不在此範圍則印出”輸入錯誤”。
n = int(input("請輸入月份 (1-12): "))

if n in (3, 4, 5):
    print("春")
elif n in (6, 7, 8):
    print("夏")
elif n in (9, 10, 11):
    print("秋")
elif n in (12, 1, 2):
    print("冬")
else:
    print("輸入錯誤")
```

```python
# 2.	選擇性敘述的練習-salary
# 輸入便利商店工讀生的工作時數，並計算其薪資。
# 60小時以內，時薪120元。
# 61~80小時，以時薪1.25倍計算。
# 81小時以上，以時薪1.5倍計算。
# 說明：薪資以累計方式計算。若工時為90小時，則薪資為60*120 + 20*120*1.25 + 10*120*1.5元。
n = int(input("輸入便利商店工讀生的工作時數:"))

if n <= 60:
    print(f"您本月薪資為{n * 120}元")
elif 61 <= n <= 80:
    print(f"您本月薪資為{7200 + (n-60) * 120 * 1.25}元")
elif 81 <= n:
    print(f"您本月薪資為{7200 + 3000 + (n-80) * 120 * 1.5}元")
else:
    print("請輸入整數數字")
```

```python
# 3.	選擇性敘述的練習-leapYear
# 輸入一西元年，如2015。判斷此年份是否為閏年。
# 提示：每四年一閏，每百年不閏，每四百年一閏。
n = int(input("今年是?"))

if (n % 4 == 0 and n % 100 != 0) or (n % 400 == 0):
    print(f"{n}是閏年")
else:
    print(f"{n}不是閏年")
```

```python
# 4.	選擇性敘述的練習-refund
# 輸入在某商店購物應付金額與實付金額。
# 實付金額小於應付金額，則印出”金額不足”。
# 實付金額等於應付金額，則印出”不必找錢”。
# 實付金額大於應付金額，則輸出找回金額最少的鈔票數和錢幣數。
# 假設幣值只有1000, 500, 100元紙鈔和50, 10, 5, 1元硬幣。
# 說明：若買了132元的商品，付1000元，應找回一張500元，三張100元，一個50元硬幣，一個10元硬幣，一個5元硬幣和三個1元硬幣。
pp, paid = map(int, input("請輸入應付金額與實付金額（以空白隔開）：").split())

if paid < pp:
    print("金額不足")
elif paid == pp:
    print("不必找錢")
else:
    change = paid - pp

    # 用 // 拿張數，用 % 拿剩餘金額
    n1000 = change // 1000
    change %= 1000  # 這等同於 change = change % 1000

    n500 = change // 500
    change %= 500

    n100 = change // 100
    change %= 100

    n50 = change // 50
    change %= 50

    n10 = change // 10
    change %= 10

    n5 = change // 5
    n1 = change % 5

    print(
        f"應找回：{n1000}張1000元，{n500}張500元，{n100}張100元，"
        f"{n50}個50元，{n10}個10元，{n5}個5元，{n1}個1元"
    )
```
